```python
# Python code to illustrate Sending mail with attachments
# from your Gmail account

# libraries to be imported
import pandas as pd
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.base import MIMEBase
from email import encoders
```

```python

fromaddr = "b18097@students.iitmandi.ac.in"

#######################################################
email_list = pd.read_excel('/home/yashwinder/Desktop/email_id.xlsx')

# getting the names and the emails
#names = email_list['NAME']
emails = email_list['EMAIL']

# iterate through the records
for i in range(len(emails)):

    toaddr = emails[i]
    
    # instance of MIMEMultipart
    msg = MIMEMultipart()

    # storing the senders email address
    msg['From'] = fromaddr

    # storing the receivers email address
    msg['To'] = toaddr
    
    msg['Cc'] = "b18046@students.iitmandi.ac.in,b18066@students.iitmandi.ac.in" 
    
    # storing the subject
    msg['Subject'] = "Request for online research project"

    # string to store the body of the mail
    body = "Hello Professor,\n\nMy name is Yashwinder Kumar. I am a fourth-year B.Tech student at Indian Institute Technology Mandi, majoring in Computer Science Engineering. I am writing this letter on behalf of my team members to ask you about research opportunities available for final year undergraduate students. We all study at IIT Mandi and majoring in Computer Science and Technology. \n\nRecently, we three did a project, where we performed statistical and descriptive analysis on the survey \"perception of risk, fear, social distancing, mask and treatment regarding COVID-19 in the US and India\" at the Indian Institute of Technology Mandi in an online semester, which is on the way of getting featured in the UNESCO Paris conference. This project will help to estimate different aspects of mental and physical problems faced by people during the COVID-19 period. \n\nWe also did some projects on Operating systems (includes sockets, threading and kernel), Networking (includes TCP, routers and switches) and Data Science during our first three years of undergraduate. We have done several courses such as Automata theory, Communication of Distributed processes, system practicum, computer architecture, Information and database systems, paradigms of programming, Data Science and Machine learning. All these courses have motivated us enough to enter into the research field that will greatly help us consolidate our knowledge and future career choice.\n\nWe are hard-working, committed and enthusiastic students who have a deep understanding of Data Science, Communication of Distributed processes and Machine learning. We have good understanding, communication and are comfortable enough to work in a team. We are extremely interested in taking project opportunities under your supervision. We are comfortable enough to work on an online project due to COVID-19 issues and also contribute to our college semester credits during the final year (2021-2022).\n\nWe believe that research allows a broader educational experience where students are able to explore the effects of applying new thought processes through study and testing. We have seen your several papers on google scholar and your work really fascinated us to work under your supervision and guidance during our last year in the college. We would appreciate the chance to talk with you about research on a topic of interest and about possible year-long undergraduate research opportunities.\n\nWe know you are very busy. We could schedule an appointment or can drop by your office hours on day and time. Please let us know if there is any other information to provide. We look forward to talking to you soon. We are attaching our resumes in the mail below. Waiting for your response.\n\nYours Sincerely,\nYashwinder Kumar\nKumar Harsh\nAnam Siddiqui"

    # attach the body with the msg instance
    msg.attach(MIMEText(body, 'plain'))

    #################################################################
    # open the file to be sent
    filename = "Yashwinder_resume.pdf"
    attachment = open("/home/yashwinder/Documents/MTP_Resume/Yashwinder_resume.pdf", "rb")

    # instance of MIMEBase and named as p
    p = MIMEBase('application', 'octet-stream')

    # To change the payload into encoded form
    p.set_payload((attachment).read())

    # encode into base64
    encoders.encode_base64(p)

    p.add_header('Content-Disposition', "attachment; filename= %s" % filename)

    # attach the instance 'p' to instance 'msg'
    msg.attach(p)
    ##################################################################
    
    #################################################################
    # open the file to be sent
    filename = "anam_siddiqui_Resume.pdf"
    attachment = open("/home/yashwinder/Documents/MTP_Resume/anam_siddiqui_Resume.pdf", "rb")

    # instance of MIMEBase and named as p
    p = MIMEBase('application', 'octet-stream')

    # To change the payload into encoded form
    p.set_payload((attachment).read())

    # encode into base64
    encoders.encode_base64(p)

    p.add_header('Content-Disposition', "attachment; filename= %s" % filename)

    # attach the instance 'p' to instance 'msg'
    msg.attach(p)
    ##################################################################
    
    #################################################################
    # open the file to be sent
    filename = "Resume_KumarHarsh.pdf"
    attachment = open("/home/yashwinder/Documents/MTP_Resume/Resume_KumarHarsh.pdf", "rb")

    # instance of MIMEBase and named as p
    p = MIMEBase('application', 'octet-stream')

    # To change the payload into encoded form
    p.set_payload((attachment).read())

    # encode into base64
    encoders.encode_base64(p)

    p.add_header('Content-Disposition', "attachment; filename= %s" % filename)

    # attach the instance 'p' to instance 'msg'
    msg.attach(p)
    ##################################################################

    # creates SMTP session
    s = smtplib.SMTP('smtp.gmail.com', 587)

    # start TLS for security
    s.starttls()

    # Authentication
    s.login(fromaddr, "Password")

    # Converts the Multipart msg into a string
    text = msg.as_string()

    # sending the mail
    s.sendmail(fromaddr, toaddr, text)

# terminating the session
s.quit()
```
